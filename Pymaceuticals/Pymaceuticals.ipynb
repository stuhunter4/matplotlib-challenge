## Observations and Insights : The best performing drug regimens were Capomulin, slightly second to Ramicane.  Both drugs yielded far lower average tumor volume measurements, as well as a much lower range of tumor size measurements.  A sample mouse's tumor treated with Capomulin began to yield positive results at roughly the middle timepoint of the study, when the tumor volume dropped dramatically below its original measurement.  This progress was not linear.  There is a strong association between weight and average tumor volume, within the Capomulin data.  The weight of a given mouse was a potential predictor of smaller sized tumors.

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse
ID_group = clean_mouse.groupby(["Mouse ID"])
last_timepoint = ID_group["Timepoint"].max()
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
timepoint_df = pd.merge(clean_mouse, last_timepoint, how="left", on=["Mouse ID", "Mouse ID"])
timepoint_df = timepoint_df.rename(columns={"Timepoint_x":"Timepoint", "Timepoint_y":"Last Timepoint"})
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drugs_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
last_time_list = []
# Locate the rows which contain mice on each drug and get the tumor volumes
for x in drugs_list:
    last_time = timepoint_df.loc[(timepoint_df["Drug Regimen"] == x) & (timepoint_df["Timepoint"] == timepoint_df["Last Timepoint"])]
    last_time_list.append(last_time)
Capomulin_list = last_time_list[0]["Tumor Volume (mm3)"]
Ramicane_list = last_time_list[1]["Tumor Volume (mm3)"]
Infubinol_list = last_time_list[2]["Tumor Volume (mm3)"]
Ceftamin_list = last_time_list[3]["Tumor Volume (mm3)"]

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
Capomulin_quartiles = Capomulin_list.quantile([.25,.5,.75])
Capomulin_lowerq = Capomulin_quartiles[0.25]
Capomulin_upperq = Capomulin_quartiles[0.75]
Capomulin_iqr = Capomulin_upperq - Capomulin_lowerq
Capomulin_lowerb = Capomulin_lowerq - (1.5*Capomulin_iqr)
Capomulin_upperb = Capomulin_upperq + (1.5*Capomulin_iqr)

Ramicane_quartiles = Ramicane_list.quantile([.25,.5,.75])
Ramicane_lowerq = Ramicane_quartiles[0.25]
Ramicane_upperq = Ramicane_quartiles[0.75]
Ramicane_iqr = Ramicane_upperq - Ramicane_lowerq
Ramicane_lowerb = Ramicane_lowerq - (1.5*Ramicane_iqr)
Ramicane_upperb = Ramicane_upperq + (1.5*Ramicane_iqr)

Infubinol_quartiles = Infubinol_list.quantile([.25,.5,.75])
Infubinol_lowerq = Infubinol_quartiles[0.25]
Infubinol_upperq = Infubinol_quartiles[0.75]
Infubinol_iqr = Infubinol_upperq - Infubinol_lowerq
Infubinol_lowerb = Infubinol_lowerq - (1.5*Infubinol_iqr)
Infubinol_upperb = Infubinol_upperq + (1.5*Infubinol_iqr)

Ceftamin_quartiles = Ceftamin_list.quantile([.25,.5,.75])
Ceftamin_lowerq = Ceftamin_quartiles[0.25]
Ceftamin_upperq = Ceftamin_quartiles[0.75]
Ceftamin_iqr = Ceftamin_upperq - Ceftamin_lowerq
Ceftamin_lowerb = Ceftamin_lowerq - (1.5*Ceftamin_iqr)
Ceftamin_upperb = Ceftamin_upperq + (1.5*Ceftamin_iqr)

# add subset 
# Determine outliers using upper and lower bounds
Capo_out_df = timepoint_df.loc[((timepoint_df["Drug Regimen"] == "Capomulin") & (timepoint_df["Timepoint"] == timepoint_df["Last Timepoint"])) & ((timepoint_df["Tumor Volume (mm3)"] > Capomulin_upperb) | (timepoint_df["Tumor Volume (mm3)"] < Capomulin_lowerb))]
Rami_out_df = timepoint_df.loc[((timepoint_df["Drug Regimen"] == "Ramicane") & (timepoint_df["Timepoint"] == timepoint_df["Last Timepoint"])) & ((timepoint_df["Tumor Volume (mm3)"] > Ramicane_upperb) | (timepoint_df["Tumor Volume (mm3)"] < Ramicane_lowerb))]
Infu_out_df = timepoint_df.loc[((timepoint_df["Drug Regimen"] == "Infubinol") & (timepoint_df["Timepoint"] == timepoint_df["Last Timepoint"])) & ((timepoint_df["Tumor Volume (mm3)"] > Infubinol_upperb) | (timepoint_df["Tumor Volume (mm3)"] < Infubinol_lowerb))]
Ceft_out_df = timepoint_df.loc[((timepoint_df["Drug Regimen"] == "Ceftamin") & (timepoint_df["Timepoint"] == timepoint_df["Last Timepoint"])) & ((timepoint_df["Tumor Volume (mm3)"] > Ceftamin_upperb) | (timepoint_df["Tumor Volume (mm3)"] < Ceftamin_lowerb))]

print(Capo_out_df[["Mouse ID", "Tumor Volume (mm3)", "Drug Regimen"]])
print(Rami_out_df[["Mouse ID", "Tumor Volume (mm3)", "Drug Regimen"]])
print(Infu_out_df[["Mouse ID", "Tumor Volume (mm3)", "Drug Regimen"]])
print(Ceft_out_df[["Mouse ID", "Tumor Volume (mm3)", "Drug Regimen"]])
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
correlation = st.pearsonr(Mouse_weight, Mouse_tumor)
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(Mouse_weight, Mouse_tumor)
line_eq = Mouse_weight * slope + intercept

plt.scatter(Mouse_weight, Mouse_tumor)
plt.plot(Mouse_weight, line_eq,"r-")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
print(f"The correlation coefficient between weight and average tumor volume is {round(correlation[0],2)}")
plt.show()
```
