## Observations and Insights 


```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
mouse_complete = pd.merge(study_results, mouse_metadata, how="left", on=["Mouse ID", "Mouse ID"])

# Display the data table for preview
mouse_complete
```

```python
# Checking the number of mice.
total_mice = len(mouse_complete["Mouse ID"].unique())
total_mice
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate = mouse_complete[mouse_complete.duplicated(["Mouse ID","Timepoint"],keep=False)]
duplicate
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
dup_mouse = mouse_complete.loc[mouse_complete["Mouse ID"] == "g989", :]
dup_mouse
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_mouse = mouse_complete[mouse_complete["Mouse ID"] != "g989"]
clean_mouse
```

```python
# Checking the number of mice in the clean DataFrame.
len(clean_mouse["Mouse ID"].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# This method is the most straighforward, creating multiple series and putting them all together at the end.
drug_group = clean_mouse.groupby(["Drug Regimen"])
drug_mean = drug_group["Tumor Volume (mm3)"].mean()
drug_median = drug_group["Tumor Volume (mm3)"].median()
drug_var = drug_group["Tumor Volume (mm3)"].var()
drug_stdev = drug_group["Tumor Volume (mm3)"].std()
drug_sem = drug_group["Tumor Volume (mm3)"].sem()
drug_summ = pd.DataFrame({"Mean": drug_mean, "Median": drug_median, "Variance": drug_var, "Standard Deviation": drug_stdev, "SEM": drug_sem})
drug_summ
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of mice for each treatment throughout the course of the study using pandas. 
drug_mice_total = drug_group["Mouse ID"].count()
drug_mice_total = drug_mice_total.sort_values(ascending=False)
drug_mice_total_plot = drug_mice_total.plot(kind="bar")
drug_mice_total_plot.set_ylabel("Number of Unique Mice Tested")
plt.show()
```

```python
# Generate a bar plot showing the total number of mice for each treatment throughout the course of the study using pyplot.
drug_mice_count = clean_mouse["Drug Regimen"].value_counts()
x_axis = range(0, len(clean_mouse["Drug Regimen"].unique()))
tick_locations = []
for x in x_axis:
    tick_locations.append(x)
    
plt.xlabel("Drug Regimen")
plt.ylabel("Number of Unique Mice Tested")

plt.bar(x_axis, drug_mice_count)
plt.xticks(tick_locations, drug_mice_total.index, rotation="vertical")
plt.show()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pandas
gender_count = clean_mouse["Sex"].value_counts()
gender_list = gender_count.keys()
gender_plot = gender_count.plot(kind="pie", y=gender_list, autopct="%1.1f%%")
plt.show()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pyplot
plt.pie(gender_count, labels=gender_list, autopct="%1.1f%%")
plt.ylabel("Sex")
plt.show()
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse
ID_group = clean_mouse.groupby(["Mouse ID"])
last_timepoint = ID_group["Timepoint"].max()
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
timepoint_df = pd.merge(clean_mouse, last_timepoint, how="left", on=["Mouse ID", "Mouse ID"])
timepoint_df = timepoint_df.rename(columns={"Timepoint_x":"Timepoint", "Timepoint_y":"Last Timepoint"})
```

```python
# Put treatments into a list for for loop (and later for plot labels)
promising_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
# Create empty list to fill with tumor vol data (for plotting)
new_list = []
for x in promising_list:
    last_time = timepoint_df.loc[(timepoint_df["Drug Regimen"] == x) & (timepoint_df["Timepoint"] == timepoint_df["Last Timepoint"])]
    new_list.append(last_time)
Capomulin_list = new_list[0]["Tumor Volume (mm3)"]
Ramicane_list = new_list[1]["Tumor Volume (mm3)"]
Infubinol_list = new_list[2]["Tumor Volume (mm3)"]
Ceftamin_list = new_list[3]["Tumor Volume (mm3)"]
# Calculate the IQR and quantitatively determine if there are any potential outliers. 
Capomulin_quartiles = Capomulin_list.quantile([.25,.5,.75])
Capomulin_lowerq = Capomulin_quartiles[0.25]
Capomulin_upperq = Capomulin_quartiles[0.75]
Capomulin_iqr = Capomulin_upperq - Capomulin_lowerq
Capomulin_lowerb = Capomulin_lowerq - (1.5*Capomulin_iqr)
Capomulin_upperb = Capomulin_upperq + (1.5*Capomulin_iqr)

total_outliers = []
for y in range(0, 4):
    outlier_data = [(new_list[y]["Tumor Volume (mm3)"] < Capomulin_lowerb) | (new_list[y]["Tumor Volume (mm3)"] > Capomulin_upperb)]
    total_outliers.append(outlier_data)
    # Locate the rows which contain mice on each drug and get the tumor volumes
total_outliers    
    
    # add subset 
    
    
    # Determine outliers using upper and lower bounds
    
```
